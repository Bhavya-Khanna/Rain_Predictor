NEXT TIME DIRECT GO FOR K MEAN CLASSIFICATION FOR TARGET VARIABLE AS RAINTODAY

```python
X= df[['Humidity3pm','Humidity9am','Cloud9am','Temp3pm','Cloud3pm','MaxTemp','Sunshine','RISK_MM','Pressure9am','Pressure3pm','WindGustSpeed','Evaporation','WindSpeed9am','WindSpeed3pm','']].values
X[0:5]
```

```python
Y= df[['RainToday']].values
Y[0:5]
```

```python
#normalize data
from sklearn import preprocessing
X= preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.95,random_state=4)
print('train set: ',X_train.shape, Y_train.shape)
print('test set: ',X_test.shape, Y_test.shape)
```

```python
#now for classification
from sklearn.neighbors import KNeighborsClassifier
```

```python
#training
k=10
neigh = KNeighborsClassifier(n_neighbors=k).fit(X_train,Y_train)
neigh
```

```python
#predicting
yhat= neigh.predict(X_test)
yhat[0:5]
```

```python
#accuracy evaluation
from sklearn import metrics
print("Train set accuracy: ", metrics.accuracy_score(Y_train,neigh.predict(X_train)))
print("Test set accuracy: ", metrics.accuracy_score(Y_test,yhat))
```
